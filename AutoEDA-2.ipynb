```python
import pandas as pd
import sweetviz as sv

```

```python
df = pd.read_csv("train.csv")

# Display the first few rows to confirm
df.head()
```

```python
# Create an analysis report
report = sv.analyze(df)

# Display the report in a notebook (if supported)
report.show_html("Sweetviz_Report.html")  # Saves the report as an HTML file

```

```python
#Comparing train and test datasets
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

# Generate a comparison report
comparison_report = sv.compare([train, "Training Data"], [test, "Test Data"])
comparison_report.show_html("Comparison_Report.html")
```
